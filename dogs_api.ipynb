```python
dogs_df = dogs_df.drop_duplicates()
dogs_df.count()
```

```python
dogs_df["Avg Max Weight"] = (dogs_df['Max Weight Male'] + dogs_df['Max Weight Female'])/2
dogs_df["Avg Max Height"] = (dogs_df['Max Height Male'] + dogs_df['Max Height Female'])/2
dogs_df["Avg Min Weight"] = (dogs_df['Min Weight Male'] + dogs_df['Min Weight Female'])/2
dogs_df["Avg Min Height"] = (dogs_df['Min Height Male'] + dogs_df['Min Height Female'])/2
dogs_df.head()
```

```python
results = stats.linregress(dogs_df['Avg Max Weight'],dogs_df['Barking'])
print(results)
```

```python
## Define a function to categorize dogs based on weight
def categorize_dog(value):
	if value <= 22:
		return "1-Small"
	elif (value > 22) & (value <= 59):
		return "2-Medium"
	elif (value >59) & (value <= 99):
		return "3-Large"
	elif value >99:
		return "4-X-large"

## Apply this new function using the pandas.map() function
dogs_df["Weight Category"] = dogs_df["Avg Max Weight"].map(categorize_dog)
```

```python
#find the differences between size and barking
dog_groupby = dogs_df.groupby(["Weight Category","Barking"])["Barking"].count()
dog_groupby
```

```python
#find the differences between size and barking
dog_groupby = dogs_df.groupby(["Weight Category"]).mean()["Barking"]
dog_groupby
```

```python
#graph for weight and barking - Do smaller dogs bark more?
dog_groupby.plot(kind = "bar", y label = "Barking", title="Size vs Barking", facecolor="darkorange", alpha=0.65, rot=45, legend=True)

#.../Images/SizeVsBarkingBarChart.png
# Save the figure

plt.savefig("../Images/SizeVsBarkingBarChart.png", bbox_inches="tight")

plt.show()
```
